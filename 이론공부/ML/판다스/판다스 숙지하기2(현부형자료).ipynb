## 데이터 다루기 

### 열단위 다루기 



*   데이터를 열(columns) 단위로 다루기

> [방법1] '데이터 프레임 이름'.iloc[:,변수 위치]

> 방법 1은 index 값을 이용하는 함수이다.

> [방법2] '데이터 프레임 이름'['변수명']

> 방법 2는 변수 명을 직접 입력하는 방법이다.

> [방법3] '데이터 프레임 이름'.loc[:,변수명]

> column을 그냥 불러오는 방법.

> 조건문에서 많이 사용되는 방법이 방법 3이다.


````
print("[방법1] '데이터 프레임 이름'.iloc[:,변수 위치]")
print(iris.iloc[:,1])
"""
[행,렬]=>[Row, Column]이고, [:,0]의 의미는 column1의 모든 데이터를 불러오겠다는 것임.
"""
print("[방법2] '데이터 프레임 이름'['변수명']")
print(iris['Id'])
print("[방법3] '데이터 프레임 이름'.loc[:,변수명]")
print(iris.loc[:,'SepalLengthCm'])
````

### 행단위 다루기 



*   데이터를 행(rows) 단위로 다루기

> [방법] '데이터 프레임 이름'.iloc[행 위치,:]

*   행열 동시에 다루기

> [방법1] '데이터 프레임 이름'iloc[행위치, 열위치]

> [방법2] '데이터 프레임 이름'[변수명][행위치]

> [방법3] '데이터 프레임 이름'[변수명].iloc[행위치]

> [방법4] '데이터 프레임 이름'.loc[:,변수명].iloc[행위치]


````
print("[방법] '데이터 프레임 이름'.iloc[행 위치,:]")
print(iris.iloc[0:6,:])
````

````
print("[방법1] '데이터 프레임 이름'iloc[행위치, 열위치]")
print(iris.iloc[0,0])
print("[방법2] '데이터 프레임 이름'[변수명][행위치]")
print(iris['SepalLengthCm'][0])
print("[방법3] '데이터 프레임 이름'[변수명].iloc[행위치]")
print(iris['Id'].iloc[0])
print("[방법4] '데이터 프레임 이름'.loc[:,변수명].iloc[행위치]")
print(iris.loc[:,'Id'].iloc[0])
````

## 선택한 열의 기초 통계 계산 



*   빈도 테이블 및 기초 통계량 계산

> [빈도 테이블:문자형 변수] '데이터 프레임 이름'[변수명].value_counts()

> [평균: 수치형 변수] '데이터 프레임 이름'[변수명].mean()

> [중앙값: 수치형 변수] '데이터 프레임 이름'[변수명].median()

> [분산: 수치형 변수] '데이터 프레임 이름'[변수명].var()

> [표준편차: 수치형 변수] '데이터 프레임 이름'[변수명].std()




````
print("[빈도 테이블:문자형 변수] '데이터 프레임 이름'[변수명].value_counts()")
print(iris['Species'].value_counts()) # 열 값의 종류별 개수
print("[평균: 수치형 변수] '데이터 프레임 이름'[변수명].mean()")
print(iris['PetalWidthCm'].mean()) # 열 값의 평균
print("[평균: 수치형 변수] '데이터 프레임 이름'[변수명].median()")
print(iris['PetalWidthCm'].median()) # 이건 평균이 아니라 중간값 같은데 아마 ddd
print("[분산: 수치형 변수] '데이터 프레임 이름'[변수명].var()")
print(iris['PetalWidthCm'].var()) 
print("[표준편차: 수치형 변수] '데이터 프레임 이름'[변수명].std()")
print(iris['PetalWidthCm'].std())
````

## 특정 조건의 행과 열 데이터 추출하기

> [방법] '데이터 프레임 이름'.loc[행 조건,:] or '데이터 프레임 이름'.loc[:,열 조건]

> [예제] SepalLengthCm 변수가 5cm 보다 큰 행 데이터 추출하기

>> iris.loc[iris['SepalLengthCm']>5,:]

>>> 여기 행 중 SepalLengthCm가 5보다 큰 모든 행을 열과 상관 없이 뽑으라는 의미



````
print(iris.loc[iris['SepalLengthCm']>5,:])
````

###  추출 된 데이터의 기초 통계량 구하기

> [Stpe1] 저장할 데이터 명 = 처리 구문

> [Step2] 저장한 데이터 명.함수명()



````
new_iris = iris.loc[iris['SepalLengthCm']>5,:]
print('범주형 변수: 빈도')
print('원래 데이터')
print(iris['Species'].value_counts())
print('추출 데이터')
print(new_iris["Species"].value_counts())


print('수치형 변수: 평균')
print('원래 데이터')
print(iris.iloc[:,1:4].apply(lambda x:x.mean(),axis=0))
print('추출 데이터')
print(new_iris.iloc[:,1:4].apply(lambda x:x.mean(),axis=0))

print('수치형 변수: 분산')
print('원래 데이터')
print(iris.iloc[:,1:4].apply(lambda x:x.std(),axis=0))
print('추출 데이터')
print(new_iris.iloc[:,1:4].apply(lambda x:x.std(),axis=0))
````

## Group by : 선택한 칼럼과 다른 칼럼에 대한 데이터 처리 
* 예시를 보느게 빠름
> [방법] '데이터 프레임 이름'.[[변수명]].groupby('데이터 프레임 이름'.[그룹 변수명])

````
print('변수 하나만 Grouping 하기')
group_iris = iris[['SepalLengthCm']].groupby(iris['Species'])
print(group_iris.mean())
print(group_iris.sum())
print(group_iris.count())
print(group_iris.size())
````

````
print('변수 두개 이상 Grouping 하기')
group_iris = iris[['SepalLengthCm','SepalWidthCm']].groupby(iris['Species'])
print(group_iris.mean())
print(group_iris.sum())
print(group_iris.count())
print(group_iris.size())
````

````
print('모든 변수 Grouping 하기')
group_iris = iris.groupby(iris['Species'])
print(group_iris.mean())
print(group_iris.sum())
print(group_iris.count())
print(group_iris.size())
````

## 데이터 합치기 : concat

> [방법] pd.concat([[데이터프레임1], [데이터 프레임2]], axis = 0 (행) or axis = 1(열)



pd.concat(objs,  # Series, DataFrame, Panel object

             axis=0,  # 0: 위+아래로 합치기, 1: 왼쪽+오른쪽으로 합치기

             join='outer', # 'outer': 합집합(union), 'inner': 교집합(intersection)

             join_axes=None, # axis=1 일 경우 특정 DataFrame의 index를 그대로 이용하려면 입력

             ignore_index=False,  # False: 기존 index 유지, True: 기존 index 무시
             keys=None, # 계층적 index 사용하려면 keys 튜플 입력

             levels=None,

             names=None, # index의 이름 부여하려면 names 튜플 입력

             verify_integrity=False, # True: index 중복 확인
             copy=True) # 복사



상세정보 여기로 : https://rfriend.tistory.com/256 [R, Python 분석과 프로그래밍의 친구 (by R Friend)]

#### 특정 칼럼값인 데이터 추출하기 

````
print('Species가 Iris-setosa 인 데이터 추출')
setosa = iris.loc[iris['Species']=='Iris-setosa',:]
setosa.head()
````

````
print('Species가 Iris-versicolor 인 데이터 추출')
versicolor = iris.loc[iris['Species']=='Iris-versicolor',:]
versicolor.head()
````

````
print("setosa 데이터와 versicolor 데이터를 행으로 합치기")
merge_by_row = pd.concat([setosa, versicolor],axis=0)
print(merge_by_row)
````

````
print("setosa 데이터와 versicolor 데이터를 열으로 합치기")
merge_by_col = pd.concat([setosa, versicolor],axis=1)
print(merge_by_col)
````

> [인덱스 초기화 방법] '데이터 프레임 이름'.reset_index(inplace = True)

> [인덱스 삽입 방법] '데이터 프레임 이름'.index('index')



````
print("drop = True 명령어를 사용할 경우")
print(setosa.reset_index(drop=True).head())
print("drop = True 명령어를 사용하지 않을 경우")
print(setosa.reset_index(drop=False).head())
````

````
setosa_head = setosa.head()
setosa_head.index = ['A','B','C','D','E']
print(setosa_head)
````

````
print("reset_index 명령어 없이 setosa 데이터와 versicolor 데이터를 행으로 합치기")
merge_by_row = pd.concat([setosa, versicolor],axis=1)
print(merge_by_row)
print(merge_by_row.iloc[0,:])
````

````
print("reset_index 명령어 후 setosa 데이터와 versicolor 데이터를 행으로 합치기")
versicolor = versicolor.reset_index(drop=True)
merge_by_row = pd.concat([setosa, versicolor],axis=1)
print(merge_by_row)
print(merge_by_row.iloc[0,:])
````
