## OpenCV Tracker
- 이동(track)하면서 입력대로 조정되는!

image.png

cv2.createTracker(track_bar, name, window_name, value, count, on_change): Tracker를 생성하는 함수

- value: 초기 값
- count: Max 값 (Min: 0)
- on_change: 값이 변경될 때 마다 호출되는 Callback 함수

cv2.getTrackerPos(track_bar, name, window_name): Tracker로부터 값을 얻어 오는 함수

### OpenCV Tracker 실습

- 코랩(matplot)으로는 화면 출력이 힘들다
- 파이참이나 vscode ㄱㄱ

```
import cv2
import numpy as np

# getTrackbarPos 이용해서 콜백함수 만들어주기
def change_color(x):
  r = cv2.getTrackbarPos("R", "Image")
  g = cv2.getTrackbarPos("G", "Image")
  b = cv2.getTrackbarPos("B", "Image")
  image[:] = [b, g, r] # opencv는 항상 b,g,r 순서이다!!
  cv2.imshow('Image', image)
  
image = np.zeros((600, 400, 3), np.uint8) # 600,400크기를 0으로 채운다
cv2.namedWindow("Image")

# createTrackbar
cv2.createTrackbar("R", "Image", 0, 255, change_color)
cv2.createTrackbar("G", "Image", 0, 255, change_color)
cv2.createTrackbar("B", "Image", 0, 255, change_color)

cv2.imshow('Image', image)
cv2.waitKey(0)
```
