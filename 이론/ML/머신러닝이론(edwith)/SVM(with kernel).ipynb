```python
from sklearn.datasets import load_breast_cancer, load_boston
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, SVR
```

# SVM을 사용한 분류(SVC)
- classfier의미에서 svc임 

```python
X,y = load_breast_cancer(return_X_y=True) # 이렇게하면 알아서 X,y에 할당됨
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)
# random_state를 특정상수로 고정하는 이유는, 그래야 각 실행마다 같은값이 나옴 ㅇㅇ
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
print(f'학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'평가 데이터 점수 : {model.score(X_test,y_test)}')
```

# SVM을 사용한 회귀(SVR) 

```python
X,y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)
model = SVR()
model.fit(X_train, y_train)
print(f'학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'평가 데이터 점수 : {model.score(X_test,y_test)}')
# 이거 지금 보스턴 데이터셋 자체에 문제있는듯. 하여간 뭐 위에랑 똑같다 ㅇㅇ
```

# 커널 기법을 사용한 분류(SVC) 

```python
X,y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)

# 커널종류 linear, poly(다항식커널), rbf(유사도)
model = SVC(kernel='linear')
model.fit(X_train, y_train)
print(f'kernel(linear) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(linear) 평가 데이터 점수 : {model.score(X_test,y_test)}')

model = SVC(kernel='poly')
model.fit(X_train, y_train)
print(f'kernel(poly) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(poly) 평가 데이터 점수 : {model.score(X_test,y_test)}')

model = SVC(kernel='rbf')
model.fit(X_train, y_train)
print(f'kernel(rbf) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(rbf) 평가 데이터 점수 : {model.score(X_test,y_test)}')
```

# 커널 기법을 사용한 회귀(SVR) 

```python
X,y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)

# 커널종류 linear, poly(다항식커널), rbf(유사도)
model = SVR(kernel='linear')
model.fit(X_train, y_train)
print(f'kernel(linear) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(linear) 평가 데이터 점수 : {model.score(X_test,y_test)}')

model = SVR(kernel='poly')
model.fit(X_train, y_train)
print(f'kernel(poly) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(poly) 평가 데이터 점수 : {model.score(X_test,y_test)}')

model = SVR(kernel='rbf')
model.fit(X_train, y_train)
print(f'kernel(rbf) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(rbf) 평가 데이터 점수 : {model.score(X_test,y_test)}')
```

# SVM 매개변수
* SVM 커널에 따라 다양한 매개변수 설정 가능
* 주로 사용하는 매개변수
  * kernel : 사용할 커널기법 종류
  * C : 하드 마진, 소프트마진 결정
    * 데이터 샘플이 다른 클래스에 놓이는 것을 허용하는지를 결정
    * 낮게 설정하면 이상치를 허용 => 과소적합
    * 높게 설정하면 이상치 없음 => 과대적합
  * gamma : 유사도 특성 관련 

```python
X,y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)


model = SVC(kernel='poly', degree=2, gamma='auto', C=0.5) # degree는 그 차수추가를 몇차로할지임 ㅇㅇ
model.fit(X_train, y_train)

print(f'kernel(poly) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(oply) 평가 데이터 점수 : {model.score(X_test,y_test)}')
```

```python
X,y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100)


model = SVC(kernel='rbf', gamma='scale', C=2) # rbf는 degree 설정이 당연히 없다
model.fit(X_train, y_train)

print(f'kernel(poly) 학습 데이터 점수 : {model.score(X_train,y_train)}')
print(f'kernel(oply) 평가 데이터 점수 : {model.score(X_test,y_test)}')
```
