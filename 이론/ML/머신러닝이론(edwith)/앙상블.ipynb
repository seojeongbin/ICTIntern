```python
from sklearn.datasets import load_iris, load_wine, load_boston, load_diabetes
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate, cross_val_score
```

* cross_validation
  * 과적합을 피하기 위해서 사용
  * 훈련 데이터 전체를 한번에 사용하지않고 일부를 남겨두고 테스트에 활용하는 것. '교차검증'
  * cross_val_score와 비슷하지만, 분할마다 훈련과 테스트에 걸린 시간을 담은 딕셔너리를 반환한다.
* cross_val_score : 걍 score 정확도 따지는건데, 좀더 잘 나눈다 뭐 그런거
  * 인자(사용될 모델, X, Y, cv=반복횟수) cv 몇겹으로 따질것이냐.
  * 큰 값이 좋은 결과를 나타냄
  * 회귀 모델에선 MSE를 얻기 위해 주로 'neg_mean_squared_error'값을 사용 (0으로 갈수록 좋은 점수)
    - 회귀는 음수로 나올 수 있다
      - scoring='neg~~'
      - 앞에 - (마이너스)

# 투표기반 분류와 회귀 

```python
from sklearn.ensemble import VotingClassifier, VotingRegressor 
```

## 투표 기반 분류 

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
```

```python
# 걍 분류 알고리즘들 불러온거임
model1 = SVC()
model2 = GaussianNB()
model3 = SGDClassifier() 
```

```python
# 분류기들을 투표에 넣는방법은 이렇게 하면된다 ㅇㅇ
vote_model = VotingClassifier(
    estimators=[('svc', model1), ('gaussian', model2), ('sgd', model3)]
) # 리스트 원소를 튜플로
```

```python
for model in (model1, model2, model3, vote_model):
  model_name1 = str(type(model)).split('.')[-1][:-2]
  model_name2 = str(type(model)).split('.')[-1]
  model_name3 = str(type(model)).split('.')
  model_name4 = str(type(model))
  model_name5 = model

print(model_name1,'\n')
print(model_name2,'\n')
print(model_name3,'\n')
print(model_name4,'\n')
print(model_name5)
```

```python
# for in 직접 튜플로 묶어줄수 있다.
for model in (model1, model2, model3, vote_model):
  model_name = str(type(model)).split('.')[-1][:-2] # 여기는 그냥 그러려니 하기.. 
  scores = cross_val_score(model, iris.data, iris.target, cv=5)
  print(f'{model_name}, Accuracy: {scores.mean()}')
```

VotingClassfier의 결과와, 가장 높은 결과인 SVC는 당연히 같은듯! 

## 투표 기반 회귀
* 보통 투표 기반 분류는 쓰는데 회귀는 잘 안쓴단다 

```python
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor, LinearRegression
```

```python
model1 = SVR()
model2 = SGDRegressor()
model3 = LinearRegression() 
```

```python
vote_model = VotingClassifier(
    estimators=[('svr', model1), ('sgd', model2), ('linear', model3)]
)
```

```python
for model in (model1, model2, model3, vote_model):
  model_name = str(type(model)).split('.')[-1][:-2]
  scores = cross_val_score(model, boston.data, boston.target, cv=5)
  print(f'{model_name}, Accuracy: {scores.mean()}')
```

이렇게 회귀는 음수로 나올 수 있다
  - scoring='neg~~'
  - 앞에 -

이 조치를 해줘야함!! (머신러닝과 무관하게 그냥 사이킷런 내용이지)

```python
for model in (model1, model2, model3, vote_model):
  model_name = str(type(model)).split('.')[-1][:-2]
  scores = -cross_val_score(model, boston.data, boston.target, cv=5, scoring='neg_mean_squared_error')
  print(f'{model_name}, Accuracy: {scores.mean()}')
```

# 배깅

```python
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
```

## 배깅을 사용한 분류 

```python
from sklearn.preprocessing import StandardScaler

base_model = make_pipeline( # 두과정을 묶어준거임. SVC모델에다가 정규화까지 하겠단거
    StandardScaler(),
    SVC()
)

bagging_model = BaggingClassifier(base_model, n_estimators=10, max_samples=0.5, max_features=0.5) # 내부 파라미터는 필요할때 찾아보고

bass_cross_val = cross_validate(estimator=base_model, X=iris.data, y=iris.target, cv=5)
print(f"{bass_cross_val['test_score'].mean()}") # 배깅 전 : svc + 정규화

bagging_cross_val = cross_validate(estimator=bagging_model, X=iris.data, y=iris.target, cv=5)
print(f"{bagging_cross_val['test_score'].mean()}") # 배깅 후 : svc + 정규화 + 배깅
```

## 배깅을 사용한 회귀 

```python
from sklearn.preprocessing import StandardScaler

base_model = make_pipeline(
    StandardScaler(),
    SVR()
)

bagging_model = BaggingRegressor(base_model, n_estimators=10, max_samples=0.5, max_features=0.5)

bass_cross_val = cross_validate(estimator=base_model, X=boston.data, y=boston.target, cv=5)
print(f"{bass_cross_val['test_score'].mean()}") 

bagging_cross_val = cross_validate(estimator=bagging_model, X=boston.data, y=boston.target, cv=5)
print(f"{bagging_cross_val['test_score'].mean()}") 
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor 
```

## AdaBoost 분류 

```python
model = make_pipeline(
    StandardScaler(),
    AdaBoostClassifier()
)
```

```python
cross_val = cross_validate(estimator=model, X=iris.data, y= iris.target, cv=5)
print(f"{cross_val['test_score'].mean()}") # 안에 작은따옴표 있으니까 겉엔 큰따옴표로.
```

## AdaBoost 회귀 

```python
model = make_pipeline(
    StandardScaler(),
    AdaBoostRegressor()
)
```

```python
cross_val = cross_validate(estimator=model, X=boston.data, y= boston.target, cv=5)
print(f"{cross_val['test_score'].mean()}") # 안에 작은따옴표 있으니까 겉엔 큰따옴표로.
```
