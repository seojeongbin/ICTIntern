## 경사하강법 

```
# y = ax + b (실제 결과값) -> y_hat = wx + b (y햇은 예측값 w는 예측 기울기)
```

```
# 1. 무작위로 {w,b} 선정
w = 1.0
b = 1.0
```

```
# 2. x에서 샘플갖고 와서 y_hat 계산해봄
y_hat = x[0] * w + b
print(y_hat)
```

```
# 3. 계산한 y_hat과 결과값 y 비교
print(y[0] - y_hat)
```

```
# 4. y와 y_hat 결과 가까워지도록 w,b 조정
w_inc = w + 0.1 # 0.1늘려서 새로운 변수에 저장
y_hat_inc = x[0] * w_inc + b # 그상태에서 y_hat 계산
print(y_hat_inc)
```

```
# 5. 모든 샘플에 적용
w_rate = (y_hat_inc - y_hat) / (w_inc-w) # 기울기 증가율
w_new = w + w_rate
```
