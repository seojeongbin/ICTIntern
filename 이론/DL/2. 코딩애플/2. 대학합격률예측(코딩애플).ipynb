## 데이터프레임 임포트

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DL/코딩애플/gpascore.csv')
data = data.dropna() # 빈칸있는애들은 다 삭제
```

```python
data.head()
```

- x데이터 : [ [1행의 입력변수값 : 1,2,3,4 ] , [7,5,4,6 ] , [ _ , _ , _ , _] , [ _ , _ , _ , _] , [ _ , _ , _ , _] ]
- y데이터 : [1행답, 2행답, ... , ]

## 데이터 분할 

```python
for i, rows in data.iterrows() : # .iterrows()는 데이터프레임에서 가로 한줄씩 뽑아주는 역할임
  print(i,rows)
```

```python
y = data['admit'].values
x = []

for i, rows in data.iterrows() :
  x.append([rows['gre'],rows['gpa'],rows['rank']])
```

## 모델준비 

```python
import tensorflow as tf 
```

image.png

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(64, activation='tanh'),
  tf.keras.layers.Dense(128, activation='tanh'),
  tf.keras.layers.Dense(1, activation='sigmoid'), # 활성화함수 종류 중 (0~1사이의 확률 예측경우) sigmoid 사용한다
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# loss도 결과가 (0~1) 사이의 분류/확률 문제에서는 'binary_crossentropy'를 사용한다
```

## 학습
- model.fit(x데이터, y데이터, 전체데이터셋 반복횟수 : 문제집 몇회독)
  - 파이썬에서 리스트를 바로 인자로 넣을 수 없으므로 넘파이로(혹은 텐서로) 바꾸어서 입력해준다 

```python
x = np.array(x)
y = np.array(y)
model.fit(x, y, epochs=10)
```

```python
model.fit(x, y, epochs=1000)
```

## 예측 
- model.predict([[예측하고픈 데이터행1],[데이터행2]..])

```python
prediction = model.predict([[750, 3.70, 3],[750, 3.70, 3]]) 
print(f'학생1과 학생2의 대학합격확률 입니다 : {prediction*100}%')
```
